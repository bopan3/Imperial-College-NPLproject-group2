## __1.3 文本格式数据繁体化为简体 (该格不需运行（耗时很长），如需运行先pip install opencc)

```python
import os, sys
import logging
from optparse import OptionParser
from opencc import OpenCC

def zh_t2s(infile, outfile):
    '''convert the traditional Chinese of infile into the simplified Chinese of outfile'''

    # read the traditional Chinese file
    t_corpus = []
    with open(infile,'r',encoding='utf-8') as f:
        for line in f:
            line = line.replace('\n', '').replace('\t','')
            t_corpus.append(line)
    #logger.info('read traditional file finished!')

    # convert the t_Chinese to s_Chinese
    cc = OpenCC('t2s')
    s_corpus = []
    for i,line in zip(range(len(t_corpus)),t_corpus):
        #if i % 1000 == 0:
            #logger.info('convert t2s with the {}/{} line'.format(i,len(t_corpus)))
        # s_corpus.append(OpenCC.convert(line))
        s_corpus.append(cc.convert(line))
    #logger.info('convert t2s finished!')

    # write the simplified Chinese into the outfile
    with open(outfile, 'w', encoding='utf-8') as f:
        for line in s_corpus:
            f.writelines(line + '\n')
    #logger.info('write the simplified file finished!')

"""
if __name__ == '__main__':
    program = os.path.basename(sys.argv[0])
    logger = logging.getLogger(program)
    logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

    logger.info('running ' + program + ' : convert Traditional Chinese to Simplified Chinese')

    parser = OptionParser()
    parser.add_option('-i','--input',dest='input_file',default='corpus.zhwiki.txt',help='traditional file')
    parser.add_option('-o','--output',dest='output_file',default='corpus.zhwiki.simplified.txt',help='simplified file')
    (options,args) = parser.parse_args()

    input_file = options.input_file
    output_file = options.output_file

    try:
        zh_t2s(infile=input_file,outfile=output_file)
        logger.info('Traditional Chinese to Simplified Chinese Finished')
    except Exception as err:
        logger.info(err)
"""





""" 如需重新运行 将这段代码去注释

for sub_dir_name in os.listdir('text'):
    if sub_dir_name[0]!='.':
        
        for fname in os.listdir('text/'+sub_dir_name):
            cur_file='text/'+sub_dir_name+'/'+fname
            ot_file="./after_t2s/"+str(sub_dir_name)+'_'+fname
            zh_t2s(cur_file, ot_file)
            #print(ot_file)
            #pass
            #print(fname)

#cur_file="./text/AA/wiki_00"
#ot_file="./after_t2s/after_t2s"
#zh_t2s(cur_file, ot_file)
"""
```

## __1.4 中文分词处理 (该格不需运行（耗时很长），如需运行先pip install jiaba)

```python
import jieba
import os, sys
import logging
def split_word(infile, outfile):
    '''do wordSplit on sentences from infile, write to outfile'''

    # read the input file
    i_corpus = []
    #print(infile)
    with open(infile,'r',encoding='utf-8') as f:
        for line in f:
            #print(line[0])
            line=line.strip()
            if len(line)!=0 and line[0]!='<' and line[0]!='※':
                line = line.replace('\n', '').replace('\t','')
                i_corpus.append(line)
    #logger.info('read traditional file finished!')

    # convert the t_Chinese to s_Chinese
    #cc = OpenCC('t2s')
    f.close()
    s_corpus = []
    for i,line in zip(range(len(i_corpus)),i_corpus):
        #if i % 1000 == 0:
            #logger.info('convert t2s with the {}/{} line'.format(i,len(t_corpus)))
        # s_corpus.append(OpenCC.convert(line))
        out_sentence=" ".join(jieba.cut(line, cut_all=False))
        s_corpus.append(out_sentence)
    #logger.info('convert t2s finished!')

    # write the simplified Chinese into the outfile
    with open(outfile, 'w', encoding='utf-8') as f:
        for line in s_corpus:
            f.writelines(line + '\n')
    #logger.info('write the simplified file finished!')
    f.close()
    
stop_for_test=1    
for file_name in os.listdir('after_t2s'):
    print("Processing the {file_num}th file... Progress:{progress}%".format(file_num=stop_for_test,progress=stop_for_test/1420*100))
    if file_name[0]!='.' and (not file_name in os.listdir('after_wordSplit')):        
            cur_file='after_t2s/'+file_name
            ot_file="./after_wordSplit/"+file_name
            #print(cur_file)
            split_word(cur_file, ot_file)
            #print(ot_file)
            #pass
            #print(fname)
    stop_for_test=stop_for_test+1
```

## __2.1 模型训练（须GPU，已训练好，不需要重复训练）

```python
# import modules & set up logging
import gensim, logging,os
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
from gensim.test.utils import common_texts, get_tmpfile
from gensim.models import Word2Vec 
#sentences = [['first', 'sentence'], ['second', 'sentence']]
# train word2vec on the two sentences
#model = gensim.models.Word2Vec(sentences, min_count=1)


class MySentences(object):
    def __init__(self, dirname):
        self.dirname = dirname
 
    def __iter__(self):
        for fname in os.listdir(self.dirname):
            if fname[0]!='.':
                for line in open(os.path.join(self.dirname, fname)):
                    yield line.split()
                
sentences = MySentences('after_wordSplit/') # a memory-friendly iterator



model = gensim.models.Word2Vec(sentences,min_count=5)
model.save("word2vec.model")
```

<font size=5>
-----cos距离与欧式距离区别的对比：  
    --词向量的每一维的值代表一个具有一定的语义和语法上解释的特征。故可以将词向量的每一维称为一个词语特征。所以两个词语如果夹角很小，代表其在每一特征上都很相似，所以两个词是相似的（cos距离）。  
    
    --两个相似的词有着相似的特征向量，但他们特征值（长度）差异很大，此时他们的欧式距离也很大。但是，对于两个方向（意思）完全相反但特征值很小的词来说（比如yes/no），由于他们长度很短所以欧式距离也会很小（欧式距离）。
    
</font>
